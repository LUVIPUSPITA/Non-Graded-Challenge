```python
import matplotlib.pyplot as plt

def plot_vector2d(vector2d, origin=[0,0 ], **options):
    return plt.arrow(origin[0], origin[1], vector2d[0], vector2d[1],
              head_width=0.2, head_length=0.3, length_includes_head=True,
              **options)
```

```python
a
```

```python
np.linalg.norm(a)
```

```python
import numpy as np

arr = np.array([17,22])

plot_vector2d(arr, color='r')
plot_vector2d(np.array([10,20]), color='g')
plt.axis([0,25,0,25])
plt.grid()
plt.show()
```

```python
np.linalg.norm(arr)
```

## 3. Plot vector ke Bidang 2D

```python
u = (2, 5)
v = (3, 1)
print(u, v)
```

```python
arr_1 = np.array([2, 5])
arr_2 = np.array([3, 1])
```

```python
plot_vector2d(arr_1, color='r')
plot_vector2d(arr_2, origin=[2, 5], color='g')
plot_vector2d(np.array([5, 6]))
plt.axis([0, 10, 0, 10])
plt.axvline(0)
plt.axhline(0)
plt.grid()
plt.show()
```
